```python
def print_affine(matrix, translation):

    out = []

    for row, trans in zip(matrix, translation):
        this = [row[0], row[1], row[2], trans]

        out.extend(this)

    print(out)

t = t2

mat = t["matrix"]
trans = t["translation"]

rescaled = np.array(mat) * np.array([0.65, 0.65, 1])
rescaled_trans = np.array(trans) * np.array([0.65, 0.65, 1])

print(mat)
print(rescaled)

print_affine(rescaled, rescaled_trans)
```

```python
def parse(tf):
    mat = np.array(tf["matrix"])
    trans = np.expand_dims(np.array(tf["translation"]), -1)

    full = np.hstack([mat, trans])

    full = np.vstack([full, [[0, 0, 0, 1]]])

    return full

base = np.array([[8, 0, 0, 0,],
                 [0, 8, 0, 0,],
                 [0, 0,   4, 0,],
                 [0, 0,   0, 1,],])

new = base.copy()

for tf in transforms:

    affine = parse(tf)
    new = np.matmul(affine, new)

out = ""

for x in new[:3, :].flatten():

    out = out + f"{x: 0.6f}, "

print(out)

new = base.copy()

for tf in transforms2:

    affine = parse(tf)
    new = np.matmul(affine, new)

out = ""

for x in new[:3, :].flatten():

    out = out + f"{x: 0.6f}, "

print(out)



```
